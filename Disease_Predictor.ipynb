Day 2 : Model Training & Evaluation

```python
cat_cols = df.select_dtypes(include='object').columns.tolist()
if 'num' in cat_cols:
   cat_cols.remove('num')
```

```python
x = df.drop('num', axis=1)
y = (df['num']>0).astype(int)
```

```python
x = pd.get_dummies(x, columns = cat_cols)
print("Final feature columns:", x.columns)
```

One-hot encoding creates new binary columns for each unique value in the original categorical columns. This is necessary because most machine learning algorithms cannot directly work with categorical data.



Day 3: Train/Test Split, Normalisation, Modeling and Model Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state = 42 )
```

new value = (x-mean)/standard deviation

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```



*   fit: It calculates the mean and standard deviation of each feature in your x_train data.
*   transform: It then scales the x_train data using the calculated means and standard deviations. The formula for scaling is (x - mean) / standard deviation.



```python
from sklearn.linear_model import LogisticRegression   #Its about Classification
```

```python
lr_model = LogisticRegression()     #Just like giving admission to a new student
lr_model.fit(x_train_scaled, y_train)    # Training Step
```

** Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import classification_report
y_pred_lr = lr_model.predict(x_test_scaled)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

Accuracy Score - 75 to 98 Considered to be good

Day 4 : Random Forest, and Feature Importance

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples')
plt.title('Confusion Matrix (Logistic Regression)')
plt.show()
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state = 42)   # n_estimators=100 means the model will build 100 decision trees
rf_model.fit(x_train_scaled, y_train)
y_pred_rf = rf_model.predict(x_test_scaled)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
cm = confusion_matrix(y_test, y_pred_rf)  # Extra Step
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples')
plt.title('Confusion Matrix (Logistic Regression)')
plt.show()
```

Feature Importance

```python
feat_imp = pd.Series(rf_model.feature_importances_, index=x.columns)   # This line creates a pandas Series where the index is the feature names
feat_imp.nlargest(10).plot(kind = 'barh')    # This selects the top 10 most important features and creates a horizontal bar plot to visualize their importance scores.
plt.title('Random Forest Feature Importance')
plt.show()
```

```python
import joblib
import pandas as pd

user_df = pd.read_csv('heart_dataset.csv')

#Getting Columns list from training dataframe
numeric_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
bool_cols = df.select_dtypes(include = 'bool').columns.tolist()

#Dropping Columns which are extra in user_df than required to avoid error
numeric_cols = [col for col in numeric_cols if col in user_df.columns]
cat_cols = [col for col in cat_cols if col in user_df.columns]
bool_cols = [col for col in bool_cols if col in user_df.columns]

#Fill the missing numeric columns, cat column & bool column
user_df[numeric_cols] = user_df[numeric_cols].fillna(user_df[numeric_cols].mean())

for col in cat_cols:
  user_df[col]=user_df[col].fillna('Unknown')

for col in bool_cols:
  user_df[col]=user_df[col].astype(int)

#One-hot encoding cat columns
user_df_encoded = pd.get_dummies(user_df,columns = cat_cols)

#Allign columns
user_df_encoded = user_df_encoded.reindex(columns=x.columns, fill_value=0)

#Scale Data
scaler = joblib.load('heart_scaler.pkl')
user_df_scaled = scaler.transform(user_df_encoded)

#Load Model
model = joblib.load('heart_disease_rf_model.pkl')
preds = model.predict(user_df_scaled)
user_df['Heart_Disease_Prediction'] = preds

#Show Result
print(user_df)
```
